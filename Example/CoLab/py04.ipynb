## 例外處理機制
* 例外並不全然是程式的邏輯錯誤
  * 例如：開啟檔案時檔案不存在、網路連線中斷、資料庫無回應
  * 當外部因素消失，程式恢復正常運作
* 使用例外處理保護程式不至於中斷
  * 捕捉這些例外，並撰寫相關的處理流程

## 例外處理
  * try / except : 捕捉發生的例外及其處理流程。
  * try / else : 未發生例外時執行流程。
  * try / finally: 不論有沒有發生例外都會處理的流程。

### 例外丟出
  * raise : 在程式碼中手動觸發或重丟捕捉到的例外。




## 例外處理語法

* try / except : 捕捉並處理 Python 引發的例外狀況。
    * try:\
        受例外機制保護的程式區塊
        區塊內若產生例外狀況會停止繼續向下執行，
        跳到 except 區塊。

    * except:\
        try 區塊產生例外時執行此區塊。
    
    * 宣告要處理的例外類別
      * 逐一比對例外型態
      * 發生的例外未被宣告時，程式仍會中斷

```python
num1 = 10
num2 = 0
nums = [1, 3, 5, 7, 9]
try:
  # print(num1/num2)
  print(num1*num3)
  print(nums[100])
except ZeroDivisionError:
  print("Error, Zero Division Error")
except (NameError, IndexError):
  print("Error, Naming Error or Index Error")
except:
  print("Error!!")
```

## try else

* else 區塊代表 try 區塊沒有產生例外時執行的區塊。
* 只有在try 區塊內的程式碼是被保護的
  * except / else 區塊內程式碼產生例外狀況時仍會造成程式中斷。
  * 可使用巢狀try結構保護


```python
num1 = 10
num2 = 0
nums = [1, 3, 5, 7, 9]

try:
  print(num1/(num2+1))
  print(num1*num2)
  print(nums[4])
except:
  print("Error")
else:
  print("Non-Error")
```

## try finaly

* finally 區塊代表不論 try 區塊執行後一定要執行的區塊。
  * 不論有沒有產生例外，例外有沒有被處理都會執行
  * 程式中產生的例外沒有被處理而導致中斷，finally 區塊會在中斷前執行。


```python
num1 = 10
num2 = 0
nums = [1, 3, 5, 7, 9]
try:
  # print(num1/num2)
  print(num1*num3)
  print(nums[100])
except ZeroDivisionError:
  print("Error, Zero Division Error")
except (NameError, IndexError):
  print("Error, Naming Error or Index Error")
except:
  print("Error!!")
finally:
  print("結束!!")
```
